```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Load Data

```python
psl_results = pd.read_csv("../results/fairness/psl/fairness_study/psl_fairness_cp.csv")
psl_results
```

```python
psl_results = pd.read_csv("../results/fairness/psl/fairness_study/psl_fairness.csv")

nmf_result = psl_results[psl_results.Fairness_Model == 'nmf'].iloc[0]

np_0 = nmf_result.copy()
np_0.loc['Fairness_Model'] = 'non_parity_nmf_retro_fit'
np_0.Fairness_Regularizer = '0.0'
psl_results = psl_results.append(np_0)

val_0 = nmf_result.copy()
val_0.loc['Fairness_Model'] = 'value_nmf_retro_fit'
val_0.Fairness_Regularizer = '0.0'
psl_results = psl_results.append(val_0)
psl_results
```

# Retrofitting Results

```python
retrofitting_points = ["0.0", "0.00001", "0.0001", "0.001", "0.01", "0.1", "1.0", "10.0", "100.0", "1000.0", "10000.0"]
```

## Non-Parity

```python
retrofitting_points = ["0.0", "0.01", "0.1", "1.0", "10.0", "100.0", "1000.0", "10000.0"]
```

```python
psl_np_retrofitting = psl_results[(psl_results.Fairness_Model == 'non_parity_nmf_retro_fit') 
                                  & (psl_results.Fairness_Regularizer.isin(retrofitting_points))
                                 & (psl_results.Wl_Method != 'BOWLSS') ]
psl_np_retrofitting = psl_np_retrofitting.sort_values(by="Fairness_Regularizer")
psl_np_retrofitting.iloc[:, 5:] = psl_np_retrofitting.iloc[:, 5:] * 5
psl_np_retrofitting
```

```python
# create figure and axis objects with subplots()
fig,ax = plt.subplots()
# make a plot
np_line = ax.plot(psl_np_retrofitting.Fairness_Regularizer, psl_np_retrofitting.non_parity_Mean, color="red", marker="o", label='Non-Parity Unfairness')
# error bars
ax.errorbar(psl_np_retrofitting.Fairness_Regularizer, psl_np_retrofitting.non_parity_Mean, 
           yerr=psl_np_retrofitting.non_parity_Standard_Deviation,
          capsize=4)
# set x-axis label
ax.set_xlabel("Fairness Regularizer",fontsize=14)
# set y-axis label
ax.set_ylabel("Non-Parity Unfairness",color="black",fontsize=14)
ax.set_ylim([max(psl_np_retrofitting.non_parity_Mean.min() - 2 * psl_np_retrofitting.non_parity_Standard_Deviation.max(), 0), 
             psl_np_retrofitting.non_parity_Mean.max() + 2 * psl_np_retrofitting.non_parity_Standard_Deviation.max()])

# twin object for two different y-axis on the sample plot
ax2=ax.twinx()
# make a plot with different y-axis using second axis object
rmse_line = ax2.plot(psl_np_retrofitting.Fairness_Regularizer, psl_np_retrofitting.Evaluator_Mean, color="blue", marker="o",
        label='RMSE')
# error bars
ax2.errorbar(psl_np_retrofitting.Fairness_Regularizer, psl_np_retrofitting.Evaluator_Mean, 
            yerr=psl_np_retrofitting.Evaluator_Standard_Deviation,
           capsize=4)
ax2.set_ylabel("RMSE", color="black", fontsize=14)
ax2.set_ylim([psl_np_retrofitting.Evaluator_Mean.min() - 2 * psl_np_retrofitting.Evaluator_Standard_Deviation.max(), 
              psl_np_retrofitting.Evaluator_Mean.max() + 2 * psl_np_retrofitting.Evaluator_Standard_Deviation.max()])

lns = np_line + rmse_line
labs = [l.get_label() for l in lns]
ax.legend(lns, labs, loc='center left')
plt.savefig('./NPvsFairnessRegularizer.png', bbox_inches='tight')
plt.show()
```

# Value

```python
retrofitting_points = ["0.0", "0.00001", "0.0001", "0.001", "0.01", "0.1", "1.0", "10.0"]
```

```python
psl_value_retrofitting = psl_results[(psl_results.Fairness_Model == 'value_nmf_retro_fit') 
                                  & (psl_results.Fairness_Regularizer.isin(retrofitting_points))
                                    & (psl_results.Wl_Method != 'BOWLSS')]

psl_value_retrofitting = psl_value_retrofitting.sort_values(by="Fairness_Regularizer")
psl_value_retrofitting.loc[:, "Fairness_Regularizer"] = psl_np_retrofitting.loc[:, "Fairness_Regularizer"].values
psl_value_retrofitting.iloc[:, 5:] = psl_value_retrofitting.iloc[:, 5:] * 5
psl_value_retrofitting
```

```python
# create figure and axis objects with subplots()
fig,ax = plt.subplots()
# make a plot
val_line = ax.plot(psl_value_retrofitting.Fairness_Regularizer, psl_value_retrofitting.value_Mean, 
                   color="red", marker="o", label='Value Unfairness')
# error bars
ax.errorbar(psl_value_retrofitting.Fairness_Regularizer, psl_value_retrofitting.value_Mean, 
            yerr=psl_value_retrofitting.value_Standard_Deviation,
           capsize=4)
# set x-axis label
ax.set_xlabel("Fairness Regularizer",fontsize=14)
# set y-axis label
ax.set_ylabel("Value Unfairness",color="black",fontsize=14)
ax.set_ylim([psl_value_retrofitting.value_Mean.min() - 2 * psl_value_retrofitting.value_Standard_Deviation.max(), 
             psl_value_retrofitting.value_Mean.max() + 2 * psl_value_retrofitting.value_Standard_Deviation.max()])

# twin object for two different y-axis on the sample plot
ax2=ax.twinx()
# make a plot with different y-axis using second axis object
rmse_line = ax2.plot(psl_value_retrofitting.Fairness_Regularizer, psl_value_retrofitting.Evaluator_Mean, 
                     color="blue", marker="o", label='RMSE')
# error bars
ax2.errorbar(psl_value_retrofitting.Fairness_Regularizer, psl_value_retrofitting.Evaluator_Mean, 
             yerr=psl_value_retrofitting.Evaluator_Standard_Deviation,
            capsize=4)
ax2.set_ylabel("RMSE", color="black", fontsize=14)
ax2.set_ylim([psl_np_retrofitting.Evaluator_Mean.min() - 2 * psl_np_retrofitting.Evaluator_Standard_Deviation.max(), 
              psl_np_retrofitting.Evaluator_Mean.max() + 2 * psl_np_retrofitting.Evaluator_Standard_Deviation.max()])
lns = val_line + rmse_line
labs = [l.get_label() for l in lns]
ax.legend(lns, labs, loc='center left')
plt.savefig('./ValuevsFairnessRegularizer.png', bbox_inches='tight')
plt.show()
```
